```
def findLargerCategory(s):
    if ('astro-ph' in s) or ('cond-mat' in s) or ('gr-qc' in s) or ('hep-' in s) or ('nlin.' in s) or ('nucl' in s) or ('-ph' in s) or ('math-ph' in s) or ('.' not in s):
        return 'physics'
    
    i = 0
    cat = ''
    while s[i] != '.':
        cat += s[i]
        i += 1
        if i >= len(s):
            print(s)
    if (cat == "q-bio"):
        return "qbio"
    if (cat == "q-fin"):
        return "qfin"
    return cat
```
